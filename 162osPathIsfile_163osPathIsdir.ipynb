```python
#162

import os

target = input('파일 여부를 확인할 경로를 입력하세요: ')

if os.path.isfile(target):
    print('[%s] 는 파일입니다.' % target)
else:
    print('[%s] 는 파일이 아닙니다.' % target)
```

```python
#163

import os

target = input('디렉터리 여부를 확인할 경로를 입력하세요: ')

if os.path.isdir(target):
    print('[%s] 는 디렉터리입니다.' % target)
else:
    print('[%s] 는 디렉터리가 아닙니다.' % target)
```
