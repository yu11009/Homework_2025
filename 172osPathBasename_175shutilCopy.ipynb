```python
#172

import os

path = input('베이스 이름을 알고 싶은 경로를 입력하세요: ')

bname = os.path.basename(path)
print('베이스 이름:', bname)
```

```python
#175

import shutil

src = input('복사할 원본 파일 이름을 입력하세요: ')
dst = input('대상 파일 이름(또는 경로)를 입력하세요: ')

try:
    shutil.copy(src, dst)
    print('[%s] 를 [%s] 로 복사했습니다.' % (src, dst))
except Exception as e:
    print(e)
```
