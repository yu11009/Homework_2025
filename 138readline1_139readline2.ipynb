```python
#138

f = open('stockcode.txt', 'r')
line_num = 1
line = f.readline()
while line:
   print('%d %s' %(line_num, line), end='')
   line = f.readline()
   line_num += 1
f.close()
```

```python
#139

f = open('stockcode.txt', 'r')
lines = f.readlines()
#print(lines)
for line_num, line in enumerate(lines):
   print('%d %s' %(line_num+1, line), end='')
f.close()
```
