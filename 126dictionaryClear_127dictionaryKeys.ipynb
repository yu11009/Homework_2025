```python
#127 사전 키 추출

names = {'Mary':10999, 'Sams':2111, 'Aimy':9788, 'Tom':20245}
ks = names.keys()
print(ks)

for k in ks:
  print('Key:%s \tValue:%d' %(k,names[k]))
```
