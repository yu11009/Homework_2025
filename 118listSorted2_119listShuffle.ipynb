```python
#118 리스트 정렬2

nl = ['Mary', 'Sams', 'Aimy', 'Tom', 'Michale', 'Bob', 'Kelly']
r1 = sorted(nl)
r2 = sorted(nl, reverse=True)
print(r1)
print(r2)
```

```python
#119 리스트 섞기

from random import shuffle

ld = list(range(1, 11))
for i in range(3):
    shuffle(ld)
    print(ld)
```
