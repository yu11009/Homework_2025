```python
#128 사전 값 추출

names = {'Mary':10999, 'Sams':2111, 'Aimy':9788, 'Tom':20245}
vals = names.values()
print(vals)

vals_list = list(vals)
ret = sum(vals_list)
print(ret)
```

```python
#128 사전 요소 모두 추출

names = {'Mary':10999, 'Sams':2111, 'Aimy':9788, 'Tom':20245}
it = names.items()
print(it)

for iterm in it:
    print(iterm)
```
