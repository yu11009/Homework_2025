```python
#152

import os

newfolder = input('새로 생성할 디렉터리 이름을 입력하세요: ')

try:
    os.mkdir(newfolder)
    print('[%s] 디렉터리를 새로 생성했습니다.' % newfolder)
except Exception as e:
    print(e)
```

```python
#153

import os

delfolder = input('삭제할 디렉터리 이름을 입력하세요: ')

try:
    os.rmdir(delfolder)
    print('[%s] 디렉터리를 삭제했습니다.' % delfolder)
except Exception as e:
    print(e)
```
