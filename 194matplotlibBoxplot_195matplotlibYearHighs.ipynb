```python
#194

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)

    highs = []

    for row in data:
        if row[4] == '':
            continue
        t = float(row[4])
        highs.append(t)

plt.boxplot(highs)
plt.show()


#195

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)

    year = input('일별 최고기온을 보고 싶은 연도를 입력하세요(예:2016): ')
    highs = []

    for row in data:
        if row[4] == '':
            continue
        if row[0].split('-')[0] == year:
            t = float(row[4])
            highs.append(t)

plt.plot(highs, 'r')
plt.show()
```
