```python
#132 문자 코드값

ch = input('문자를 1개 입력하세요: ')
if len(ch) != 0:
   ch = ch[0]
   chv = ord(ch)
   print('문자: %s \t코드값: %d [%s]' %(ch, chv, hex(chv)))
```

```python
#133 코드값 대응 문자

val = input('문자 코드값을 입력하세요: ')
val = int(val)
try:
   ch = chr(val)
   print('코드값: %d [%s], 문자: %s' %(val, hex(val), ch))
except ValueError:
   print('입력한 <%d>에 대한 문자가 존재하지 않습니다!' %val)
```
