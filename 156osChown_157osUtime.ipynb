```python
#156

import os

target = input('소유권을 변경할 파일 이름을 입력하세요: ')
uid = int(input('소유자 UID를 입력하세요: '))
gid = int(input('그룹 GID를 입력하세요: '))

try:
    os.chown(target, uid, gid)
    print('[%s] 파일의 소유권을 uid=%d, gid=%d 로 변경했습니다.' % (target, uid, gid))
except Exception as e:
    print(e)
```

```python
#157

import os, time

target = input('수정시간을 변경할 파일 이름을 입력하세요: ')
new_mtime = int(input('새로운 수정시간(유닉스 타임스탬프)을 입력하세요: '))

try:
    atime = os.path.getatime(target)
    os.utime(target, (atime, new_mtime))
    print('[%s] 파일의 수정시간을 %s 로 변경했습니다.' %
          (target, time.ctime(new_mtime)))
except Exception as e:
    print(e)
```
