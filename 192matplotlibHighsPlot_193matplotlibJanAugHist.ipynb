```python
#192

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)

    highs = []

    for row in data:
        if row[4] == '':
            continue
        t = float(row[4])
        highs.append(t)

plt.plot(highs, 'r')
plt.show()


#193

import csv
import matplotlib.pyplot as plt

with open('seoul.csv', 'r') as f:
    data = csv.reader(f)
    header = next(data)

    jan = []
    aug = []

    for row in data:
        if row[4] == '':
            continue
        month = row[0].split('-')[1]
        t = float(row[4])
        if month == '01':
            jan.append(t)
        elif month == '08':
            aug.append(t)

plt.hist(jan, bins=100, color='blue', alpha=0.5)
plt.hist(aug, bins=100, color='red', alpha=0.5)
plt.show()
```
