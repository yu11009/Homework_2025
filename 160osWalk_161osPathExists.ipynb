```python
#160

import os

top = input('탐색을 시작할 최상위 디렉터리 경로를 입력하세요: ')

for path, dirs, files in os.walk(top):
    print('경로:', path)
    print('디렉터리:', dirs)
    print('파일:', files)
    print('------------------------')
```

```python
#161

import os

target = input('존재 여부를 확인할 경로를 입력하세요: ')

if os.path.exists(target):
    print('[%s] 는 존재합니다.' % target)
else:
    print('[%s] 는 존재하지 않습니다.' % target)
```
