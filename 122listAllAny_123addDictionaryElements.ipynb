```python
#123 사전 요소 추가

solar1 = ['태양', '수성', '금성','지구', '화성', '목성', '토성', '토성', '천왕성', '해왕성']
solar2 = ['Sun','Mercury','Venus','Earth','Mars','Jupiter','Saturn','Uranus','Neptune']
solardick = {}
for i,k in enumerate(solar1):
  val = solar2[i]
  solardick[k] = val

print(solardick)
```
